```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.datasets import mnist

import tensorflow as tf
import numpy as np
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
#X_train, X_test = X_train / 255.0, X_test / 255.0 #Normalization

print('MNIST Dataset Shape:')
print('X_train: ' + str(X_train.shape))
print('Y_train: ' + str(Y_train.shape))
print('X_test:  '  + str(X_test.shape))
print('Y_test:  '  + str(Y_test.shape))
```

# 1 MLP

```python
model = Sequential()
model.add(Flatten())
model.add(Dense(60, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)

#model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
model.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=20, batch_size=256)

model.summary()
```

```python
loss, acc = model.evaluate(X_test, Y_test, verbose=1)
```

# 2 Dropout layer

## Dropout(0.2)

```python
model = Sequential()
model.add(Flatten())
model.add(Dense(60, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))

loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)

model.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=20, batch_size=256)

model.summary()

loss, acc = model.evaluate(X_test, Y_test, verbose=1)
```

## Dropout(0.6)

```python
model = Sequential()
model.add(Flatten())
model.add(Dense(60, activation='relu'))
model.add(Dropout(0.6))
model.add(Dense(10, activation='softmax'))

loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)

model.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=20, batch_size=256)

model.summary()

loss, acc = model.evaluate(X_test, Y_test, verbose=1)
```

# 3 CNN

```python
#https://keras.io/examples/vision/mnist_convnet/

X_train1 = X_train.astype("float32")
X_test1 = X_test.astype("float32")

X_train1 = np.expand_dims(X_train1, axis=-1)
X_test1 = np.expand_dims(X_test1, axis=-1)

#Y_train1 = keras.utils.to_categorical(Y_train, 10) # Converts a class vector (integers) to binary class matrix.(convert to one-hot coding) -> cross entry loss
#Y_test1 = keras.utils.to_categorical(Y_test, 10)


model = Sequential([
    Conv2D(32,3, activation='relu', input_shape=(28,28,1)),
    #https://towardsdatascience.com/batch-normalization-in-practice-an-example-with-keras-and-tensorflow-2-0-b1ec28bde96f
    BatchNormalization(), # Normalizes a tensor by mean and variance 
    keras.layers.MaxPooling2D(pool_size=(3, 3)),

    Flatten(),
    Dense(64, activation='relu'),
    Dense(10,activation='softmax')  # activation change
])
model.summary()

# loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False) #We expect labels to be provided as integers
# loss_fn = tf.keras.losses.CategoricalCrossentropy(from_logits=False) #If you want to provide labels using one-hot representation, please use CategoricalCrossentropy

model.compile(optimizer='adam', loss="sparse_categorical_crossentropy", metrics=['accuracy'])

model.fit(X_train1, Y_train, epochs=2, batch_size=2000)



loss, acc = model.evaluate(X_test1, Y_test, verbose=1)
```

## Feature map

```python
# https://machinelearningmastery.com/how-to-visualize-filters-and-feature-maps-in-convolutional-neural-networks/
# plot feature map of first conv layer for given image
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.models import Model
from matplotlib import pyplot
from numpy import expand_dims
# load the model
model = VGG16()
# redefine model to output right after the first hidden layer
model = Model(inputs=model.inputs, outputs=model.layers[2].output)
model.summary()
# load the image with the required shape
img = tf.keras.utils.get_file(origin='https://images.unsplash.com/photo-1532974297617-c0f05fe48bff?ixlib=rb-1.2.1&ixid=eyJhcHBfaWQiOjEyMDd9&auto=format&fit=crop&w=700&q=80',fname='car.jpg')
img = load_img(img, target_size=(224, 224))
pyplot.imshow(img)
pyplot.show()
#img = load_img('hi.png', target_size=(224, 224))
# convert the image to an array
img = img_to_array(img)
# expand dimensions so that it represents a single 'sample'
img = expand_dims(img, axis=0)
# prepare the image (e.g. scale pixel values for the vgg)
img = preprocess_input(img) # Preprocesses a tensor or Numpy array encoding a batch of images.
# get feature map for first hidden layer
feature_maps = model.predict(img)
# plot all 64 maps in an 8x8 squares
square = 8
ix = 1
for _ in range(square):
	for _ in range(square):
		# specify subplot and turn of axis
		ax = pyplot.subplot(square, square, ix)
		ax.set_xticks([])
		ax.set_yticks([])
		# plot filter channel in grayscale
		pyplot.imshow(feature_maps[0, :, :, ix-1])
		ix += 1
# show the figure
pyplot.show()

'''
# see with better resolution
for i in range(64):
	# plot filter channel in grayscale
	pyplot.imshow(feature_maps[0, :, :, i], cmap='gray')
	pyplot.show()
'''
```
